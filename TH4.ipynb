```python
#W4A3
#Viết chương trình tính giai thừa của một số nguyên n cho trước (n!=1×2×…×n) với 0 < n < 100
try:
    n=int(input('Nhập số nguyên n:'))
    giaithua=1
    for i in range(1,n+1):
        giaithua*=i
    print('Giai thừa của n là',giaithua)
except ValueError:
    print('Bạn nhập dữ liệu sai rồi')
```

```python
#W4A4
#Viết chương trình nhập vào một số nguyên n và in ra màn hình số chữ số (trừ dấu) của số đó. Không dùng kiểu dữ liệu string.
try:
    n=int(input('Nhập số nguyên n:'))
    n=abs(n)
    if n==0:
        count=1
    else:
        count=0
        for _ in range(10000):    
            if n>0:
                count+=1
                n//=10
        print(count)
except ValueError:
    print('Nhập sai')
```

```python
try:
    n = int(input('Nhập số nguyên n: '))
    n = abs(n)
    if n == 0:
        count = 1
    else:
        count = 0
        while n > 0:
            count += 1
            n //= 10
    print(count)
except ValueError:
    print('Nhập sai, vui lòng nhập một số nguyên.')
```

```python
#W4A12
#Một người có tài khoản tiết kiệm ở ngân hàng và gửi vào X đồng với lãi suất là 0.7% mỗi tháng. Viết chương trình tính số tiền sau N tháng người ấy rút được (cảgốc và lãi, bỏ qua phần lẻ thập phân). 
x=float(input())
n=int(input())
for _ in range(1,n+1):
    x+=(1+0.007)*x
print(int(x))
```

```python
#W4A13
try:
    a,b=map(int,input().split())
    ua=0
    for i in range (1,a):
        if a%i==0:
            ua+=i
    ub=0
    for k in range (1,b):
        if b%k==0:
            ub+=k
    if ua==b and ub==a:
        print('true')
    else:
        print('false')
except ValueError:
    print('Sai')
```
