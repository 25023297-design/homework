```python
#W7A3:
def sxnb(dayso):
  n=len(dayso)
  for i in range(n):
    swapped=False
    for j in range(0,n-i-1):
      if dayso[j]>dayso[j+1]:
        dayso[j],dayso[j+1]=dayso[j+1],dayso[j]
        swapped=True
    if not swapped:
      break
  return dayso
dayso=list(map(int,input().split()))
print(sxnb(dayso))
```

```python
#W7A3:
def sxc(dayso):
  n=len(dayso)
  for i in range(n):
    min_idx=i
    for j in range(i+1,n):
      if dayso[j]<dayso[min_idx]:
        min_idx=j
    dayso[i],dayso[min_idx]=dayso[min_idx],dayso[i]
  return dayso
dayso=list(map(int,input().split()))
print(sxc(dayso))
```

```python
#W7A4:
def xhnhieunhat(danhsach):
  if not danhsach:
    return None
  dem={}
  for phantu in danhsach:
    if phantu in dem:
      dem[phantu]+=1
    else:
      dem[phantu]=1
  phantu_max=None
  soluong_max=0
  for phantu in danhsach:
    if dem[phantu]>soluong_max:
      phantu_max=phantu
      soluong_max=dem[phantu]
  return phantu_max,soluong_max
danhsach=list(map(int,input().split()))
phantu,solan=xhnhieunhat(danhsach)
print(f'{phantu} xuat hien nhieu nhat, som nhat, {solan} lần')
```

```python
#W7A5:
def capso(danhsach,x):
  ket_qua=[]
  for i in range(len(danhsach)):
    for j in range(i+1,len(danhsach)):
      tong=danhsach[i]+danhsach[j]
      if tong==x:
        ket_qua.append((danhsach[i],danhsach[j]))
  return len(ket_qua)
danhsach=list(map(int,input().split()))
x=int(input())
print(capso(danhsach,x))
```

```python
#W7A7:
nums=list(map(int,input().split()))
n=len(nums)
dp=[1]*n
for i in range(n):
  for j in range(i):
    if nums[j]<nums[i]:
      dp[i]=max(dp[i],dp[j]+1)
print(max(dp))
```

```python
#W7A8:
n=int(input())
intervals=[]
for _ in range(n):
  l,r=map(int,input().split())
  intervals.append([l,r])
m=int(input())
queries=[]
for _ in range(m):
  q=int(input())
  queries.append(q)
intervals.sort(key=lambda x:x[1]-x[0])
result=[]
for q in queries:
  found=False
  for interval in intervals:
    if interval[0]<=q<=interval[1]:
      found=True
      result.append(interval[1]-interval[0])
      break
  if not found:
    result.append(-1)
print(result)
```

```python
#W7A10:
nums=list(map(int,input().split()))
m=int(input())
result=[]
for i in range(len(nums)):
  for j in range(i+1,len(nums)):
    a=sum(nums[i:j])%m
    result.append(a)
print(max(result))
```
