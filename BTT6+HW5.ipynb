```python
#W6A9:
d1=input().split()
dic1={}
for i in range(0,len(d1),2):
  key=d1[i]
  value=d1[i+1]
  dic1[key]=value
d2=input().split()
dic2={}
for i in range(0,len(d2),2):
  key=d2[i]
  value=d2[i+1]
  dic2[key]=value
result_dic={}
for key in dic1:
  if key in dic2:
    result_dic[key]=int(dic1[key])+int(dic2[key])
  else:
    result_dic[key]=int(dic1[key])
for key in dic2:
  if key not in dic1:
    result_dic[key]=int(dic2[key])
print(result_dic)
```
